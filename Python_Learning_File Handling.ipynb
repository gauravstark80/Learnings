```python
#circular linked list
class Node:
    def __init__(self,data,next=None):
        self.data = data
        self.next = next
class List:
    def __init__(self,head=None):
        self.head = head

    def printlist(self, first=None):
        self.first= first
        first=self.head
        while(first.next):
            print(first.data)
            if first.next==self.head:
                break
            first=first.next

    
obj = List()
obj.head=Node(5)
second = Node(10)
third = Node(15)

obj.head.next = second
second.next = third
third.next = obj.head
obj.printlist()
```

```python
#doubly linked list
class Node:
    def __init__(self, data=None, next=None, prev=None):
        self.data = data
        self.next = next
        self.prev = prev
    
class List:
    def __init__(self, head=None):
        self.head=head
    
    def printlist(self, first=None):
        self.first = first
        first=self.head
        while(first):
            print(first.data)
            first = first.next

obj=List()
obj.head=Node(10)
second = Node(20)
third = Node(30)

obj.head.next = second
second.next = third
third.next = None

obj.head.prev = None
second.prev = obj.head
third.prev = second

obj.printlist()
```

#STACKS AND QUEUES

>> Implements Stacks and Queues in Python
>> CRUD Operations on Stacks and Queues

Stacks and Queues are Linear Data Structures that can be differentiated on the basis of how they perform insertion and deletion operation as it can be seen in the diagram over here the stacks use the approach of Last In First Out (LIFO) for insertion and deletion. The insertion and deletion operations are known as pop and push in the case of stacks and the queues follow the approach of First In First Out (FIFO) and also the operations and methods are known as enqueuing and dequeuing.

Implementing Stacks and Queues in Python :
>> List in Python
>> Python Array
>> Deque from Collections Package


```python
#implement stack in Python  - Using Lists
my_list = ["Sunday","Monday","Tuesday","Wednesday","Thursday","Friday"]
my_stack = []
print(my_list)

#adding elements in the stack
for i in range(0,len(my_list)):
    my_stack.append(my_list[i])
    print(my_stack)

#removing elements from the stack
for i in range(0,len(my_stack)):
    my_stack.pop()
    print(my_stack)
    
```

```python
#updating elements in a stack
my_stack = [1,2,3,4,5,6,7,8,9,20]
def update(data, value):
    for i in range(0, len(my_stack)):
        if my_stack[i] == data:
            my_stack[i] = value
    return my_stack

update(20,10)
```

```python
#implement stacks in Python - Using deque from the in-built collections package
from collections import deque

my_stack = deque()
my_list = ["Sunday", "Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday"]

#inserting elements in the stack
for i in my_list:
    my_stack.append(i)
    print(my_stack)
print("\n")

#reading the elements from the stack
for i in range(0,len(my_stack)):
    my_stack.pop()
    print(my_stack)

#deleting the elements from the stack
for i in range(0,len(my_stack)):
    my_stack.pop()
    print(my_stack)
```

```python
#updating the elements in stack implemented using deque
stack = deque([1,2,3,4,5])
def update(data, value):
    ind = stack.index(data)
    stack[ind]=value
    return stack

update(4,3)
```

```python
#sorting a stack implemented using deque
stack_deq = deque([5,3,17,22,12,34,23,25])
stack_lis = [5,3,17,22,12,34,23,25]

#sorting a stack implemented using deque
sample = []
def sort_deq():
    for i in range(0,len(stack_deq)):
        x = stack_deq.pop()
        sample.append(x)
    sample.sort()
    for i in sample:
        stack_deq.append(i)
        
    return stack_deq

def sort_lis():
    stack_lis.sort()
    return stack_lis

#sort_deq()
sort_lis()
```

What are Queues?

```python
#implement queue in python - Using Lists
my_list = ["Sunday", "Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday"]
my_queue = []
print(my_list)

#adding elements in the queue
for i in range(0,len(my_list)):
    my_queue.append(my_list[i])
    print(my_queue)

#removing elements from the queue
for i in range(0,len(my_list)):
    my_queue.pop(0)
    print(my_queue)
```

```python
#updating queue implemented using the lists
my_queue = [1,2,3,20,5,6,7,8,9,20]
def update(data, value):
    for i in range(0, len(my_queue)):
        if my_queue[i] == data:
            my_queue[i] = value
    return my_queue

update(20,10)
my_queue
```

```python
#implement queue in python - Using Deque
my_list = ["Sunday", "Monday", "Tuesday", "Wednesday","Thursday","Friday", "Saturday"]
my_queue = deque()

#inserting elements using append()
for i in my_list:
    my_queue.append(i)
    print(my_queue)
print("\n")

#deleting elements using popleft()
for i in range(0,len(my_queue)):
    my_queue.popleft()
    print(my_queue)
print("\n")

#inserting elements using appendleft()
for i in my_list:
    my_queue.appendleft(i)
    print(my_queue)
print("\n")

#removing elements using pop()
for i in range(0, len(my_queue)):
    print(my_queue)
    my_queue.pop()
```
